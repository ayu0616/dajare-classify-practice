```python
import MeCab
```

```python
tagger = MeCab.Tagger("-Ochasen -d /opt/homebrew/lib/mecab/dic/mecab-ipadic-neologd")
```

```python
with open("./rashoumon.txt", "r") as f:
    text = f.read()
```

```python
res = tagger.parse(text)
kana = "".join([line.split("\t")[1] for line in res.splitlines()[:-1]])
```

```python
class Dajare:
    def __init__(self, s1: int, e1: int, s2: int, e2: int):
        self.s1 = s1
        self.e1 = e1
        self.s2 = s2
        self.e2 = e2

    @property
    def len(self):
        return self.e1 - self.s1 + 1

    def __repr__(self) -> str:
        return f"first: ({self.s1}, {self.e1}), second: ({self.s2}, {self.e2})"
```

```python
dajas: list[Dajare] = []
for i in range(len(kana) - 1):
    for j in range(i + 1, len(kana) - 1):
        if kana[i : i + 2] == kana[j : j + 2] and "ã€‚" not in kana[i : i + 2]:
            dajas.append(Dajare(i, i + 1, j, j + 1))
```

```python
merged_dajas: list[Dajare] = []
while dajas:
    daja = dajas[0]
    dajas = dajas[1:]
    continuous = [d for d in dajas if daja.e1 == d.s1 and daja.e2 == d.s2]
    if continuous:
        daja.e1 = continuous[0].e1
        daja.e2 = continuous[0].e2
        dajas.append(daja)
        dajas.remove(continuous[0])
    else:
        merged_dajas.append(daja)
```

```python
for d in merged_dajas:
    if d.e1 - d.s1 < 3:
        continue
    print(d)
    print(kana[d.s1 : d.e1 + 1], kana[d.s2 : d.e2 + 1])
    print()
```
